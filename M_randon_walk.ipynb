```python
from datetime import date
from pandas_datareader import data as pdr
import statsmodels.formula.api as smf
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import pandas as pd
import numpy as np
```

```python
today= str(date.today())
month=12
N= 252 #os dados de Brent são diarios
anos= 2 #anos de propagação da previsao
dates=pd.date_range(date.today(),periods=N*anos)
```

```python
rand= pd.DataFrame(2*np.random.randint(2, size=(N*anos,1000))-1, index=dates)
rand.iloc[0,:]=0
rw=rand.cumsum()
```

```python
plt.plot(rw)
plt.xlabel('Dates')
plt.ylabel('S_n')
```
